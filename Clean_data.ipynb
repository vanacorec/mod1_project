```python
import pandas as pd
import json

semi_clean_data = pd.read_csv("all_results_clean.csv")



def string_to_json(string_input):
    return json.loads(string_input.replace("'", "\""))


semi_clean_data['coordinates'] = semi_clean_data['coordinates'].apply(string_to_json)
semi_clean_data['categories'] = semi_clean_data['categories'].apply(string_to_json)
```
